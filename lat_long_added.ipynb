```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
```

```python
#reading table from wikipedia
df=pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
#df1=(df[0],inplace=True)
df1=df[0]
df1 
```

```python
# remove all rows where Borough column has value Not assigned

indexnames=df1[(df1['Borough']=='Not assigned')].index
df1.drop(indexnames,inplace=True)


```

```python
#grouping neighbourhood based on postcode

df1.sort_values('Postcode',ascending=True)
df1=df1.groupby('Postcode').agg({
                             'Neighbourhood': ', '.join, 
                             'Borough':'first' }).reset_index()
df1
#df2=df1[['Postcode']]
#df2.drop_duplicates()
```

```python
#replacing 'Not assigned' value in Neighbourhood column with Borough column's value

df1.Neighbourhood[df1['Neighbourhood']=='Not assigned'] = df1['Borough']
df1
```

```python
#dataframe dimensions
df1.shape
```

```python
df2=pd.read_csv('https://cocl.us/Geospatial_data')
df2
```

```python
new_df = pd.merge(df1, df2,  how='right', left_on=['Postcode'], right_on = ['Postal Code'])
new_df.drop(['Postal Code'],axis=1,inplace=True)
new_df
```
